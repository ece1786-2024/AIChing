```python
from agents import marking, fortune_telling
from hexagram_generator import get_alter_list, get_relations
import pandas as pd
data = pd.read_csv('../data/longest_A.csv', sep='\t')
```

```python
experiment_results = []
```

```python
import os
records = data.to_dict('records')
store_location = '../data/scoring_cleaned'
os.makedirs(store_location, exist_ok=True)
for record in records:
    question = record['Q']
    primary_id = record['Primary_id']
    alter_id = record['Alter_id']
    gan_zhi = record['Day_Stem']
    A = record['A']
    example_id = record['id']
    alter_list = get_alter_list(primary_id, alter_id)
    predict = fortune_telling(question, primary_id, gan_zhi, alter_list)
    predict_answer = predict.chat_history[-2]['content']
    score_df, explaination = marking(predict_answer, A)
    score_df.to_csv(f'{store_location}/{example_id}.csv', index=False)
    score = score_df['Score'].sum()
    total_score = score_df['Full Score'].sum()
    experiment_results.append({
        'example_id': example_id,
        'total_score': total_score,
        'score': score,
        'is_open': record['open'],
    })
    with open(f'{store_location}/{example_id}_log.md', 'w') as f:
        for log in predict.chat_history:
            f.write(f"{log['name']}: \n")
            f.write(f"{log['content']}\n")
        f.write(f"Standard Answer: {A}\n")
    with open(f'{store_location}/{example_id}_scoring.md', 'w') as f:
        f.write(explaination)
```

```python
result_df = pd.DataFrame(experiment_results)
print(result_df)
result_df.to_csv(f'{store_location}/experiment_results.csv', index=False)
print(result_df[result_df['is_open'] == 1]['score'].mean())
print(result_df[result_df['is_open'] == 0]['score'].mean())
```
